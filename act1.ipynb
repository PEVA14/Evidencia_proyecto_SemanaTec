```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = np.zeros((200,200), dtype=np.uint8)
img2 = np.zeros((200,200), dtype=np.uint8)

cv2.rectangle(img1, (50,50), (150,150), 255, -1)
cv2.circle(img2, (100,100), 60, 255, -1)

and_img = cv2.bitwise_and(img1, img2)
or_img = cv2.bitwise_or(img1, img2)
xor_img = cv2.bitwise_xor(img1, img2)
not_img = cv2.bitwise_not(img1)

plt.figure()
plt.subplot(2,3,1)
plt.title('Original Square'), 
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(2,3,2)
plt.title('Original Circle'),
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(2,3,3)
plt.title('AND Operation'),
plt.imshow(and_img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,4)
plt.title('OR Operation'),
plt.imshow(or_img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title('XOR Operation'),
plt.imshow(xor_img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,6)
plt.title('NOT Operation'),
plt.imshow(not_img, cmap='gray')
plt.axis('off')
```

```python
img = cv2.imread('act1.jpg')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_bandw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define yellow range in HSV
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([35, 255, 255])

# Create a mask for yellow
mask_yellow_hsv = cv2.inRange(img_hsv, lower_yellow, upper_yellow)

# Apply the mask to get only yellow parts
result_yellow_hsv = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_yellow_hsv)

lower_yellow_rgb = np.array([180, 180, 0])
upper_yellow_rgb = np.array([255, 255, 100])
mask_yellow_rgb = cv2.inRange(img_rgb, lower_yellow_rgb, upper_yellow_rgb)

img_bandw_3ch = cv2.cvtColor(img_bandw, cv2.COLOR_GRAY2RGB)

mask_yellow_reverse = cv2.bitwise_not(mask_yellow_rgb)

background_gray = cv2.bitwise_and(img_bandw_3ch, img_bandw_3ch, mask=mask_yellow_reverse)

fore_yellow = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_yellow_rgb)

img_yellow_filtered = cv2.add(img_bandw_3ch, fore_yellow)

# Define purple range in HSV
lower_purple = np.array([125, 100, 50])
upper_purple = np.array([160, 255, 255])

# Create mask
mask_purple = cv2.inRange(img_hsv, lower_purple, upper_purple)

 # Apply mask
result_purple = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_purple)

# Combine the masks first
combined_mask = cv2.bitwise_or(mask_yellow_hsv, mask_purple)

# 2. Invert the mask
inverse_mask = cv2.bitwise_not(combined_mask)

# 3. Apply inverse mask to grayscale image (keep grayscale where color isn't selected)
background_gray = cv2.bitwise_and(img_bandw_3ch, img_bandw_3ch, mask=inverse_mask)

foreground_color = cv2.bitwise_and(img_rgb, img_rgb, mask=combined_mask)

# 5. Combine them
img_filtered = cv2.add(background_gray, foreground_color)

plt.figure()
plt.subplot(2,3,1)
plt.title('Original Image'),
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2,3,2)
plt.title('HSV Image'),
plt.imshow(img_hsv)
plt.axis('off')

plt.subplot(2,3,3)
plt.title('Yellow Parts HSV'),
plt.imshow(result_yellow_hsv)
plt.axis('off')

plt.subplot(2,3,4)
plt.title('With yellow filtered'),
plt.imshow(img_yellow_filtered, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title('Purple Parts HSV'),
plt.imshow(result_purple)
plt.axis('off')

plt.subplot(2,3,6)
plt.title('Filtered Image'),
plt.imshow(img_filtered, cmap='gray')
plt.axis('off')
```

```python
# ========================================
#  FILTRO SEPIA CORRECTO (TONOS AMARILLOS)
# ========================================


def save_img(img):
    save_select = input("Do you wish to save the image?\n1. Yes\n2. No\n> ")

    if save_select == "1":
        filename = input("Enter a filename (with .jpg or .png extension):\n> ")
        success = cv2.imwrite(filename, img)
        if success:
            print(f" Image saved as '{filename}'")
        else:
            print(" Failed to save image. Check path or filename.")
    elif save_select == "2":
        print(" Image not saved.")
    else:
        print(" Invalid option. Image not saved.")
    return

def apply_sepia_filter(img):
  # Convertir a float y normalizar para operaciones ms precisas
  img = np.array(img, dtype=np.float32) / 255.0

  # --- 2. Matriz SEPIA ADAPTADA a OpenCV (BGR) ---
  # OJO: esta matriz est en orden BGR, no RGB
  sepia_bgr = np.array([
      [0.131, 0.534, 0.272],
      [0.168, 0.686, 0.349],
      [0.189, 0.769, 0.393]
  ])

  # --- 3. Aplicar el filtro sepia ---
  sepia = cv2.transform(img, sepia_bgr)

  # Limitar valores [0,1] y regresar a 8 bits
  sepia = np.clip(sepia, 0, 1)
  sepia = (sepia * 255).astype(np.uint8)

  # --- 4. Mostrar resultado (convertido a RGB para matplotlib) ---
  sepia_rgb = cv2.cvtColor(sepia, cv2.COLOR_BGR2RGB)

  plt.imshow(sepia_rgb)
  plt.title("Filtro SEPIA clido")
  plt.axis("off")
  plt.show()
  save_img(sepia_rgb)
  return

def apply_negative_filter(img):
  img_neg = cv2.bitwise_not(img)
  plt.imshow(cv2.cvtColor(img_neg, cv2.COLOR_BGR2RGB))
  plt.title("Filtro Negativo")
  plt.axis("off")
  plt.show()
  print("\n Now back to the terminal. Watch for the next prompt below!\n")
  save_img(img_neg)

def apply_popart_filter(img):

  # --- 1. Cargar la imagen ---
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  img_resize = cv2.resize(img, (400, 400))  # tamao uniforme

  # --- 2. Convertir a escala de grises ---
  gray = cv2.cvtColor(img_resize, cv2.COLOR_RGB2GRAY)

  # --- 3. Ecualizar contraste y normalizar ---
  gray_eq = cv2.equalizeHist(gray)
  gray_norm = gray_eq / 255.0

  # --- 4. Crear distintas combinaciones de colores fuertes ---
  # Usamos diferentes mezclas de canales para cada cuadro
  def apply_pop_color(gray_img, color):
      g = np.clip(gray_img * 255, 0, 255).astype(np.uint8)
      color_map = np.zeros((g.shape[0], g.shape[1], 3), dtype=np.uint8)
      
      if color == 'pink':
          color_map[:,:,0] = g        # R
          color_map[:,:,1] = (255 - g)//2  # G
          color_map[:,:,2] = (255 - g)     # B
      elif color == 'green':
          color_map[:,:,0] = (255 - g)//2
          color_map[:,:,1] = g
          color_map[:,:,2] = (255 - g)
      elif color == 'blue':
          color_map[:,:,0] = (255 - g)
          color_map[:,:,1] = (255 - g)//2
          color_map[:,:,2] = g
      elif color == 'yellow':
          color_map[:,:,0] = g
          color_map[:,:,1] = g
          color_map[:,:,2] = (255 - g)//4
      elif color == 'orange':
          color_map[:,:,0] = g
          color_map[:,:,1] = (g*0.5 + 100).clip(0,255)
          color_map[:,:,2] = (255 - g)//2
      elif color == 'purple':
          color_map[:,:,0] = (g*0.6 + 50).clip(0,255)
          color_map[:,:,1] = (255 - g)//3
          color_map[:,:,2] = g
      else:
          color_map[:,:,0] = g
          color_map[:,:,1] = g
          color_map[:,:,2] = g
      return color_map

  # --- 5. Generar 4 versiones con diferentes tonos ---
  img1 = apply_pop_color(gray_norm, 'pink')
  img2 = apply_pop_color(gray_norm, 'green')
  img3 = apply_pop_color(gray_norm, 'yellow')
  img4 = apply_pop_color(gray_norm, 'blue')

  # --- 6. Combinar en una cuadrcula 2x2 estilo Warhol ---
  top = np.hstack((img1, img2))
  bottom = np.hstack((img3, img4))
  pop_art = np.vstack((top, bottom))

  # --- 7. Mostrar resultado ---
  plt.figure(figsize=(8,8))
  plt.imshow(pop_art)
  plt.title("Filtro POP-ART ")
  plt.axis("off")
  plt.show()
  save_img(pop_art)

def apply_bw_filter(img):
  bw_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  plt.imshow(bw_image, cmap="gray")
  plt.title("Filtro B&W")
  plt.axis("off")
  plt.show()
  save_img(bw_image)

def apply_warm_filter(img):

  # --- 2. Aplicar efecto clido en BGR ---
  warm = img.astype(np.float32)

  # En BGR: canal 0 = azul, canal 2 = rojo
  warm[:,:,0] = np.clip(warm[:,:,0] * 0.9, 0, 255)   # Azul 
  warm[:,:,2] = np.clip(warm[:,:,2] * 1.2, 0, 255)   # Rojo 

  warm = warm.astype(np.uint8)

  # --- 3. Convertir a RGB solo para mostrar ---
  warm_rgb = cv2.cvtColor(warm, cv2.COLOR_BGR2RGB)
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  plt.imshow(warm_rgb)
  plt.title("Filtro Clido ")
  plt.axis("off")
  plt.show()
  save_img(warm_rgb)

def apply_cold_filter(img):

  # --- 2. Aplicar efecto fro en BGR ---
  cool = img.astype(np.float32)

  # En BGR: canal 0 = azul, canal 2 = rojo
  cool[:,:,0] = np.clip(cool[:,:,0] * 1.2, 0, 255)   # Azul 
  cool[:,:,2] = np.clip(cool[:,:,2] * 0.9, 0, 255)   # Rojo 

  cool = cool.astype(np.uint8)

  # --- 3. Convertir a RGB solo para mostrar ---
  cool_rgb = cv2.cvtColor(cool, cv2.COLOR_BGR2RGB)
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  # --- 4. Mostrar comparacin ---
  plt.imshow(cool_rgb)
  plt.title("Filtro Fro ")
  plt.axis("off")
  plt.show()
  save_img(cool_rgb)

def apply_drawing_filter(img):
  # --- 4 Convertir a escala de grises ---
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  # --- 5 Invertir los tonos ---
  inv = 255 - gray

  # --- 6 Aplicar desenfoque para suavizar la inversin ---
  blur = cv2.GaussianBlur(inv, (25, 25), 0)

  # --- 7 Combinar la imagen original con el desenfoque invertido ---
  # Esta divisin produce un efecto tipo lpiz
  sketch = cv2.divide(gray, 255 - blur, scale=256)

  # --- 8 Mostrar resultado ---
  plt.imshow(sketch, cmap='gray')
  plt.title("Filtro Sketch  (Dibujo a lpiz)")
  plt.axis("off")
  plt.show()

  # --- 9 (Opcional) Comparar con la imagen original ---
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


  plt.imshow(sketch, cmap='gray')
  plt.title("Sketch ")
  plt.axis("off")
  plt.show()
  save_img(sketch)


path = input("\n\nPlease write the route of the photo you wish to edit: ")
img = cv2.imread(path)

while True:
    choice = input(
        "What filter should we apply?\n"
        "1. Sepia\n"
        "2. Negative\n"
        "3. Pop-Art\n"
        "4. B&W\n"
        "5. Warm Filter\n"
        "6. Cold Filter\n"
        "7. Drawing\n"
        "8. Exit\n"
        "Your choice: "
    )

    if choice == "1":
        apply_sepia_filter(img)
    elif choice == "2":
        apply_negative_filter(img)
    elif choice == "3":
        apply_popart_filter(img)
    elif choice == "4":
        apply_bw_filter(img)
    elif choice == "5":
        apply_warm_filter(img)
    elif choice == "6":
        apply_cold_filter(img)
    elif choice == "7":
        apply_drawing_filter(img)
    elif choice == "8":
        print(" Thanks for using our program!!")
        break
    else:
        print(" Invalid choice. Please try again.")
print("Thanks for using our program!!")



```
